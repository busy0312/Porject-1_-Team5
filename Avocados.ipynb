```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# File to Load
file_to_load = "Resources/avocado.csv"

# Read Avocado File and store into Pandas data frame
avocado_data = pd.read_csv(file_to_load)

# Preview of the DataFrame

avocado_data=pd.DataFrame(avocado_data)
```

```python
delete_row = avocado_data[avocado_data["region"]=='TotalUS'].index
avocado_data= avocado_data.drop(delete_row)
avocado_data=avocado_data.drop(columns='Unnamed: 0',axis=1)
avocado_data
```

```python
avocado_data=avocado_data.rename(
                            columns={"4046":"Hass Small",
                                     "4225":"Hass Large",
                                     "4770":"All sizes",
                                     'AveragePrice':'Average Price',
                                     'type':'Type',
                                     'year':'Year',
                                     'region':'Region'
                                    
                                    })

avocado_data['Total Sales']=''
Total=avocado_data['Average Price']*avocado_data['Total Volume']
avocado_data['Total Sales']=Total
avocado_data
```

```python
data15=avocado_data[avocado_data['Year']==2015].groupby('Region')
data16=avocado_data[avocado_data['Year']==2016].groupby('Region')
data17=avocado_data[avocado_data['Year']==2017].groupby('Region')
data18=avocado_data[avocado_data['Year']==2018].groupby('Region')
small5=data15['Hass Small'].max()
small6=data16['Hass Small'].max()
small7=data17['Hass Small'].max()
small8=data18['Hass Small'].max()
hasssmall_sum=pd.DataFrame({"2015": small5,
                            "2016":small6,
                              "2017":small7,
                              "2018":small8
                              })
hasssmall_sum
```

```python
large5=data15['Hass Large'].max()
large6=data16['Hass Large'].max()
large7=data17['Hass Large'].max()
large8=data18['Hass Large'].max()
hasslarge_sum=pd.DataFrame({"2015": large5,
                              "2016": large6,
                              "2017":large7,
                              "2018":large8,})
hasslarge_sum
```

```python
all5=data15['All sizes'].max()
all6=data16['All sizes'].max()
all7=data17['All sizes'].max()
all8=data18['All sizes'].max()
allsizes_sum=pd.DataFrame({"2015": all5,
                              "2016": all6,
                              "2017":all7,
                              "2018":all8,})
allsizes_sum
```

```python
x_axis = np.arange(len(allsizes_sum))
```

```python

plt.figure(figsize=(16,3))
plt.bar(x_axis, allsizes_sum["2015"], color='r', alpha=0.5, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, allsizes_sum.index, rotation="vertical")
```
